```python
#%pip install python-docx
import docx 
from docx import Document
from docx.shared import Inches
import ipywidgets as widgets
from ipywidgets import interact, interactive, fixed, interact_manual
import time 
from jupyter_ui_poll import ui_events
from IPython.display import display

#%pip install jupyter-ui-poll
# Import docx NOT python-docx 

from docx.shared import Pt 
  
# Create an instance of a word document 
doc = docx.Document() 
  
```

```python
##create a button to add the authors
text = widgets.Text(
    value='',
    placeholder='Author(s)',
    description='String:',
    disabled=False
)

button_clicked = False
def on_click(b):
    global button_clicked
    button_clicked = True
        # Create a button widget
button = widgets.Button(description="Submit")
button.on_click(on_click)
display(text, button)


    # Wait for user to press the button
with ui_events() as poll:
    while button_clicked is False:
        poll(1)        
        time.sleep(.1)

    # Get the selected value from the widget
author=text.value
```

```python
##create a button to make a title
text = widgets.Text(
    value='',
    placeholder='Title',
    description='String:',
    disabled=False
)

button_clicked = False
def on_click(b):
    global button_clicked
    button_clicked = True
        # Create a button widget
button = widgets.Button(description="Submit")
button.on_click(on_click)
display(text, button)


    # Wait for user to press the button
with ui_events() as poll:
    while button_clicked is False:
        poll(1)        
        time.sleep(.1)

    # Get the selected value from the widget
title=text.value
```

```python
# Add Title to the document  
doc.add_heading(title, 0) 
doc.add_heading(author, 2)
```

```python
##create a button to accept text for project goal
text = widgets.Text(
    value='',
    placeholder='What was the goal of the project?',
    description='String:',
    disabled=False
)


button_clicked = False
def on_click(b):
    global button_clicked
    button_clicked = True
        # Create a button widget
button = widgets.Button(description="Submit")
button.on_click(on_click)
display(text, button)


    # Wait for user to press the button
with ui_events() as poll:
    while button_clicked is False:
        poll(10)          # React to UI events (upto 10 at a time)
        time.sleep(.1)

    # Get the selected value from the widget
goal=text.value
```

```python
# Adding paragraph with Increased font size 
doc.add_heading('Project Goal', 3) 
para = doc.add_paragraph(
    goal) 
```

```python
#create button to accept number of steps for the project
w=widgets.BoundedIntText(
    value=7,
    min=0,
    max=10,
    step=1,
    description='Steps?',
    disabled=False
)
button_clicked = False
def on_click(b):
    global button_clicked
    button_clicked = True
        # Create a button widget
button = widgets.Button(description="Submit")
button.on_click(on_click)
display(w, button)
with ui_events() as poll:
    while button_clicked is False:
        poll(10)         
        time.sleep(.1)

        # Get the selected value from the widget
listnum=w.value


```

```python
#create button to accept text for each step of the project
steps=[]
for i in range(0, listnum):
    text = widgets.Text(
    value='',
    description='String:',
    disabled=False
)
    
    button_clicked = False
    def on_click(b):
        global button_clicked
        button_clicked = True
        # Create a button widget
    button = widgets.Button(description="Next Step")
    button.on_click(on_click)
    display(text, button)


    # Wait for user to press the button
    with ui_events() as poll:
        while button_clicked is False:
            poll(10)          
            time.sleep(.1)

    # Get the selected value from the widget
    steps.append(text.value)
    


```

```python
#accept text for Summary/Next Steps
text = widgets.Text(
    value='',
    placeholder='Summary/Next Steps',
    description='String:',
    disabled=False
)


button_clicked = False
def on_click(b):
    global button_clicked
    button_clicked = True
        # Create a button widget
button = widgets.Button(description="Submit")
button.on_click(on_click)
display(text, button)


    # Wait for user to press the button
with ui_events() as poll:
    while button_clicked is False:
        poll(10)          
        time.sleep(.1)

    # Get the selected value from the widget
conclusion=text.value
```

```python

#add the project steps and the summary/next steps

doc.add_heading('Steps', 3) 
# Adding list of style name 'List Number' 
for i in steps:
    doc.add_paragraph(i, 
                  style='List Number') 
doc.add_heading('Summary/Next Steps', 3) 
doc.add_paragraph( 
    conclusion) 
  
# save the document
doc.save('example.docx')
```
